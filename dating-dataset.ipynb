```python
df1=df.drop(['expected_num_interested_in_me','expected_num_matches','shared_interests_o','shared_interests_partner','ambitous_o','ambition_partner','has_null'], axis=1)

print("After delete columns >5% missing values\n")
print("Numbers of rows: ", df1.shape[0])
print("Numbers of columns: ", df1.shape[1])

# Eliminamos todas las filas con valores nulos
df2 = df1.dropna(how='any')

print("\nAfter delete all empty rows\n")
print("Numbers of rows: ", df2.shape[0])
print("Numbers of columns: ", df2.shape[1])


print("\nPorcentage reduction data\n")
print("Reduction of rows: ", 100*(1-(df2.shape[0]/df.shape[0])))
print("Reduction of columns: ", 100*(1-(df2.shape[1]/df.shape[1])))
```

```python
suma_categorias = df2["match"].value_counts()
print(suma_categorias)
```

```python
# Convertir las variables categóricas en variables numéricas usando dummies
df2 = pd.get_dummies(df2)
df2.head(200)
```

```python
import pandas as pd
from sklearn.model_selection import cross_val_score
from sklearn.linear_model import LogisticRegression
import warnings
warnings.filterwarnings('ignore')

# Separar los datos en características (X) y etiquetas (y)
df2_X= df2.drop("match_b'0'", axis=1)
df2_X= df2.drop("match_b'1'", axis=1)

X1 = df2_X
X2 = df2_X[["age"]]
y = df2[["match_b'0'"]]

from imblearn.over_sampling import RandomOverSampler

# Aplicar el sobremuestreo aleatorio a la clase minoritaria
ros = RandomOverSampler()
X_resampled, y_resampled = ros.fit_resample(X1, y)

ros = RandomOverSampler()
X2_resampled, y2_resampled = ros.fit_resample(X2, y)

print(len(X2))
print(len(X2_resampled))

# Crear un modelo de regresión logística
model = LogisticRegression()

# Evaluar el modelo con validación cruzada
scores = cross_val_score(model, X_resampled, y_resampled, cv=5)

# Imprimir los resultados
print('Resultados de validación cruzada:', scores)
print('Exactitud media:', scores.mean())

# Crear un modelo de regresión logística
model2 = LogisticRegression()

# Evaluar el modelo con validación cruzada
scores2 = cross_val_score(model2, X2_resampled, y2_resampled, cv=5)

# Imprimir los resultados
print('Resultados de validación cruzada:', scores2)
print('Exactitud media:', scores2.mean())
```

### SVM

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm

# Create SVM with Gaussian RBF kernel
svm = svm.SVC(kernel='rbf')

# Fit the SVM to the data
svm.fit(X2_resampled, y2_resampled)

# Evaluar el modelo con validación cruzada
scores2 = cross_val_score(svm, X2_resampled, y2_resampled, cv=5)

print('Resultados de validación cruzada X2:', scores2)
print('Exactitud media:', scores2.mean())
```

```python
from sklearn.model_selection import GridSearchCV

# Crear diccionario de hiperparámetros para la búsqueda de rejilla
param_grid = {'gamma': [0.01, 0.1, 1]}

# Crear objeto de búsqueda de rejilla
grid_search = GridSearchCV(svm, param_grid, cv=5)

# Ejecutar la búsqueda de rejilla en los datos
grid_search.fit(X2_resampled, y2_resampled)

# Imprimir los mejores hiperparámetros encontrados
print('Mejores hiperparámetros:', grid_search.best_params_)

# Evaluar el modelo con validación cruzada utilizando los mejores hiperparámetros
scores2 = cross_val_score(grid_search.best_estimator_, X2_resampled, y2_resampled, cv=5)

# Imprimir los resultados de la validación cruzada
print('Resultados de validación cruzada X2:', scores2)
print('Exactitud media:', scores2.mean())
```
